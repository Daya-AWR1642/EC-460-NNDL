```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('Advertising Budget and Sales.csv')
```

```python
data1 = data.iloc[:,1:5]
```

```python
data1
```

```python
data1.describe()
```

```python
cor = data1.corr()
plt.figure(figsize = (4,4))
sns.heatmap(cor,annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
sns.pairplot(data1)
```

```python
x_data1 = np.array([data.iloc[:,1:4]])
```

```python
x_data1
```

```python
x_data1.shape
```

```python
x_data = x_data1.reshape(200,3)
```

```python
x_data.shape
```

```python
y_data = np.array([data.iloc[:,4]])
```

```python
y_data
```

```python
y_data.shape
```

```python
y_data = y_data.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
x_norm = sc1.fit_transform(x_data)
y_norm = sc1.fit_transform(y_data)
```

```python
x_norm
```

```python
y_norm
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x_norm,y_norm,test_size = 0.2,random_state=42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

### Gradient Descendent Algorithms 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

### Model Evaluation Using Visualization
#### Regression Plot

```python
fig,axes = plt.subplots(1,3,figsize = (18,6))
sns.regplot(x = 'TV Ad Budget ($)',y = 'Sales ($)',data = data1,ax = axes[0])
sns.regplot(x = 'Radio Ad Budget ($)',y = 'Sales ($)',data = data1,ax = axes[1])
sns.regplot(x = 'Newspaper Ad Budget ($)',y = 'Sales ($)',data = data1,ax = axes[2])

# for ax in axes:
#     ax.set_ylim(0,)
    
axes[0].set_title('TV Ad Budget vs. Sales')
axes[1].set_title('Radio Ad Budget vs. Sales')
axes[2].set_title('Newspaper Ad Budget vs. Sales')
plt.tight_layout()  
plt.show()
```

### Residual Plot

```python
fig,axes = plt.subplots(1,3,figsize = (18,6))
sns.residplot(x = 'TV Ad Budget ($)', y = 'Sales ($)',data = data1, ax = axes[0])
sns.residplot(x = 'Radio Ad Budget ($)', y = 'Sales ($)',data = data1, ax = axes[1])
sns.residplot(x = 'Newspaper Ad Budget ($)', y = 'Sales ($)',data = data1,ax = axes[2])
```

### Distribution Plot

```python
plt.figure(figsize=(6, 6))
col = 'Sales ($)'
sns.kdeplot(data=data1[col], color="r", label="Actual Value")
sns.kdeplot(y_pred, color="b", label="Fitted Values")
plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Density')
plt.legend()
```
