### Pandas for Merging or Concatenating DataFrames
Let's first creat three different DataFrames with similar column names  but different index labels i.e

```python
import pandas as pd
# List of Tuples
data1 = [('Jack',34,'Sydney',5),
        ('Riti', 31,'Delhi',7),
        ('Aadi',46,'New York',11)]

# List of Tuples
data2 = [('Mohit',34,'Tokyo',11),
        ('Veena',31,'London',10),
        ('Shaun',36,'Las Vegas',12)]

# List of Tuples
data3 = [('Mark',47,'Mumbai',13),
        ('Jose',43,'Yokohama',14),
        ('Ramu',49,'Paris',15)]

# creating the Dataframe  object  from list of tuples
firstDf = pd.DataFrame(data1,
                      columns = ['Name','age','place','experience'],
                      index = ['a','b','c'])

print(firstDf)

secondDf = pd.DataFrame(data2,
                      columns = ['Name','age','place','experience'],
                      index = ['d','e','f'])

print(secondDf)

thirdDf = pd.DataFrame(data3,
                      columns = ['Name','age','place','experience'],
                      index = ['g','h','i'])

print(thirdDf)
```

```python
df = pd.concat([firstDf,secondDf,thirdDf])
df
```

### Pandas - Grouping the data
- Creating a grouping organizes the data and return a groupby object
    - grouped  = obj.groupby(key)
    - grouped = obj.groupby(key,axis = 1)
    - grouped = obj.groupby([key1,key2])

```python
df = pd.DataFrame({'A':['foo','bar','foo','bar','foo','bar','foo','foo'],
                  'B':['one','one','two','three','two','two','one','three'],
                  'C':np.random.randn(8),
                  'D':np.random.randn(8)})
df
```

```python
grouped = df.groupby('A')
grouped
```
