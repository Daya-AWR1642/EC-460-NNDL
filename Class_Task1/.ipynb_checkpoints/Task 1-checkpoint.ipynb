## Q1. (a) Write python code for plotting the following activation functions for the input x in the range of -20 to 20.    (a) Sigmoid (b) TanH (c) Softsign (d) ReLU (e) Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = 1/(1+np.exp(-x))
print(y)

plt.plot(x,y)
plt.title("Sigmoid")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
print(y)

plt.plot(x,y)
plt.title("TanH")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = x /(1+np.abs(x))
print(y)

plt.plot(x,y)
plt.title("Softsign")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = np.maximum(0, x)
print(y)

plt.plot(x,y)
plt.title("ReLU")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)

y = np.exp(x)/ np.exp(x).sum()

plt.plot(x,y)
plt.title("Softmax")
plt.show()
```

## Q1. (b)Write python code for the derivative of the following activation functions and their plots for the input x in the range of -20 to 20.          (a) Sigmoid (b) TanH (c) Softsign (d) ReLU

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = 1/(1+np.exp(-x))
z = y*(1-y)
print(z)

plt.plot(x,z)
plt.title("derivative of the sigmoid")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
z = 1 - y*y
print(z)

plt.plot(x,z)
plt.title("TanH")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20)
y = x /(1+np.abs(x))

y1 = y.sum()

z = (y/y1)**2 / x**2
print(z)

plt.plot(x,z)
plt.title("Softsign")
plt.show()
```
